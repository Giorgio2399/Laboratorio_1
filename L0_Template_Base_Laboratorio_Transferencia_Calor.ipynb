# 2. Estado del Arte

<div class="alert alert-block alert-info">
    <b>1. Conducción de Calor</b>
La conducción es el mecanismo de transferencia de calor en donde la energía de laspartículas más energéticas de una sustancia se transfiere hacia las adyacentes menos energéticas como resultado  de interacciones entre esas partículas.  La  conducción sepuede dar tanto en gases, líquidos  como  en  sólidos.  En  el  caso  de  estos  últimos,  la conducción es producto de las vibraciones moleculares y del transporte de energía porparte de los electrones libres. La rapidez o razón de conducción de calor a través de un medio va a depender de laconfiguración   geométrica   de   este,   su   espesor,   el material   del   que   este   hecho   y   la diferencia de temperatura a través dicho medio. Esto es descrito por medio de la Ley deFourier de la conducción de calor:
    
\begin{align*}
\frac{dQ}{dt}= - k A  \frac{dT}{dx}\\
\frac{\Delta Q}{\Delta t}= - k A  \frac{\Delta T}{L}
\end{align*}

La resistencia termica se opone al paso del flujo de calor. 

$$
R_{cond} = \frac{L}{kA} \;\;\;\;\;\; \dot{Q} = \frac{\Delta T}{R_{total}}
$$

## Sección A - 10 W

````python
# Importación de librerias 
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('darkgrid')
sns.set_context("paper", font_scale=1.5, rc={"lines.linewidth": 2.5})

# Declaración de vectores donde Y es la Temperatura y X es la posición de los sensores

#Valores de la Tabla 3
Y=[44.716, 43.501, 43.103, 43.030]
X=[0, 0.01, 0.02, 0.03]

# Método de Mínimos Cuadrados 

Y=np.array(Y)
X=np.array(X)
n=len(X)

# Sumatorias y Medias
Ym=np.mean(Y)
Xm=np.mean(X)
sumx=np.sum(X)
sumy=np.sum(Y)
sumxy=np.sum(X*Y)
sumx2=np.sum(X**2)
sumy2=np.sum(Y**2)

# Coefincietes m y b

m=np.round((n*sumxy-sumx*sumy)/(n*sumx2-sumx**2),2)
b=np.round(Ym-m*Xm,2)

# Ecuación Obtenida

x=sym.Symbol('x')
f=m*x+b

fx=sym.lambdify(x,f)
fi=fx(X)


#Coeficiente de correlación y de determinación 
r=(n*sumxy-sumx*sumy)/((np.sqrt(n*sumx2-sumx**2))*(np.sqrt(n*sumy2-sumy**2)))
r2=r**2
                       
# Salida
print('------------')
print('RESULTADOS')
print('------------')
print('T(x)=',f)
print('Coeficiente de Correlación r=',r)
print('Coeficiente de Determicación r^2=',r2)

                       
# Comandos para la Gráfica por metodo de mínimos cuadrados

plt.figure(figsize=(12, 8), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(X,Y,'o',label='Datos') 
plt.xlabel('Posición de los Sensores (m)')
plt.ylabel('Temperatura (°C)')
plt.plot(X,fi,color='green',label=f)

# Lineas de error
for i in range(0,n,1):
    Ymin=np.min([Y[i],fi[i]])
    Ymax=np.max([Y[i],fi[i]])
    plt.vlines(X[i],Ymin,Ymax,color='blue',linestyle='dotted')

plt.legend()
plt.title('Temperatura (°C) vs Posición (m)')
plt.show()
````
